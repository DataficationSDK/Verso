# Polyglot Import Example

Demonstrates `#!import` to bring in shared utilities from another notebook.
This syntax is compatible with both Polyglot Notebooks and Verso.

```
#!import ./shared-utilities.ipynb
```

```
// Use the imported conversion helpers
Console.WriteLine($"100°C = {CelsiusToFahrenheit(100)}°F");
Console.WriteLine($"72°F = {FahrenheitToCelsius(72):F1}°C");
Console.WriteLine($"42 km = {KmToMiles(42):F1} miles");
```

```
// Use the imported statistical helpers and dataset
var temps = cityTemperatures.Values.ToList();
Console.WriteLine($"Mean temperature: {Mean(temps):F1}°C");
Console.WriteLine($"Median temperature: {Median(temps):F1}°C");

// Convert all to Fahrenheit using the imported helper
foreach (var (city, tempC) in cityTemperatures)
{
    Console.WriteLine($"  {city}: {tempC}°C / {CelsiusToFahrenheit(tempC):F1}°F");
}
```
